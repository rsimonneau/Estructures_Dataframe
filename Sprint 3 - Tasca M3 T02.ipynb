# Variables

## Identificaió del registre
- Primera columna: posició del registre

## Data i horari del vol
- Year: Any en que s'ha realitzat el vol
- Month: Mes en que s'ha realitzat el vol
- DayofMonth: Día del mes en que s'ha realitzat el vol
- DayOfWeeK: Día de la setmana en que s'ha realitzat el vol (nº)
- DepTime: Hora de sortida actual
- CRSDepTime: Hora de sortida planificada
- ArrTime: Hora d'arribada actual
- CRSArrTime: Hora d'arribada planificada

## Identificació de l'aerolinia, vol i avió
- UniqueCarrier: codi únic de l'aerolínia
- FlightNum: número de vol
- TailNum: identificació de l'avió (matrícula)

## Temps de vol
- ActualElapsedTime: temps actual del vol entre porta de sortida i porta d'arribada (minuts)
- CRSElapsedTime: temps estimat del vol entre porta de sortida i porta d'arribada (minuts)
- AirTime: temps de vol (minuts)
- ArrDelay: Diferencia en minuts entre l'hora estimada d'arribada i la real (si el vol arriba abans de l'hora estimada, número negatiu en minuts)
- DepDelay: Diferencia en minuts entre l'hora estimada de sortida i la real (si el vol surt abans de l'hora estimada, número negatiu en minuts)

## Ruta del vol
- Origin: Codi IATA de l'aeroport d'origen
- Dest: Codi IATA de l'aeroport de destí
- Distance: Distancia entre aeroports (milles)

## Temps de les operacions d'enlairament i d'aterratge
- TaxiIn: temps desde aterratge fins arribada a la porta d'embarcament (minuts)
- TaxiOut: temps desde porta d'embarcament fins l'enlairament (minuts)

## Cancel·lació, codi de cancel·lació i desviació
- Cancelled: Si el vol s'ha cancel·lat o no. 1 = Sí, 0 = No
- CancellationCode: codi segons causa de cancel·lació. (A = Aerolinia, B = Metereologia extrema, C = Sistema Nacional d'Aviació (tràfic aeri, meteorologia no extrema, operacions aeroportuaries, etc.), D = Seguretat)
- Diverted: Si el vol s'ha desviat o no. 1 = Sí, 0 = No

## Retards
- CarrierDelay:Retard aerolinia (minuts) (NaN: Not a Number)
- WeatherDelay: Retard meteorologia (minuts)
- NASDelay: Retard Sistema Nacional d'Aviació (minuts)
- SecurityDelay: Retard Seguretat (minuts)
- LateAircraftDelay: Retard per el retard del vol anterior (minuts)

 1. Anàlisi de la diferencia (minuts) entre l'hora de sortida, estimada i real, per aerolínies, per determinar quina aerolínia acumula més retrasos en la sortida

```python
# Data frame amb columnes "UniqueCarrier", "DepDelay"


import pandas as pd

DelayedFlights = pd.read_csv('DelayedFlights_Reduit.csv')

Diferencia_hora_sortida = DelayedFlights[['UniqueCarrier','DepDelay']]

Diferencia_hora_sortida.sort_values(['UniqueCarrier', 'DepDelay'])

```

```python
# Sumem las diferencies en minuts per companyia aerolinia

Diferencia_hora_sortida.groupby('UniqueCarrier').sum()
```

```python
# Ordenem els retrasos acumalats per aerolinea de major a menor.

Diferencia_hora_sortida.groupby('UniqueCarrier').sum().sort_values('DepDelay', ascending = False)
```

A priori, havent analitzat la variable ***DepDelay*** del dataframe de 200.000 resgitres, s'observa que la aerolínea que acumula més retrasos en la sortida del vols, és la aerolínea WN i la que menys retrasos acomula és la aerolínea AQ. No obstannt caldrà anlitzar i contrastar-ho amb el número de vols realitzat per cada aerolínia.

```python
# Calculem el número de vols per aerolínea

Diferencia_hora_sortida[('UniqueCarrier')].value_counts()
```

Havent calculat el número de vols per aerolínea, dels registres que tenim al dataframe, observem que efectivament la aerolínia que acumula menys retrasos és la que menys vols te registrats i la que mes retrasos acumula, és la que te més vols regitrats. Per tant l'anàlisis inicial no sería del tot correcte. Mirem de calcular la mitjana dels minuts acumulats entre les hores de sortida estimades i reals per aerolínia.

```python
Diferencia_hora_sortida.groupby('UniqueCarrier').median().sort_values('DepDelay')
```

Analitzant la mitjana dels minuts acumulats entre les hores de sortida estimades i reals per aerolínia, podriem determinar quina de les aerolínies és mes propensa a patir retrasos. Observem que la aerolínia WN que en l'analisi preliminar presentava major retràs acumulat, te un retràs mitjà de 20 minuts, i que per tant no sería la aerolinia que presenta mes retrasos en l'hora de sortida como podria interpretarse inicialment.

```python
# Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

# Nova columna "Velocitat Mitjana" en "miles / hour"

import pandas as pd
import numpy as np

DelayedFlights = pd.read_csv('DelayedFlights_Reduit.csv')

DelayedFlights['velocitat_mitjana'] = DelayedFlights['Distance'] / (DelayedFlights['AirTime'] / 60)

pd.options.display.max_columns = None

DelayedFlights
```

```python
# Nova columna "Retard?"

    # Si l'hora d'arribada actual és major a l'hora d'arribada planificada, es produeix retard
        # Si ArrTime > CRSArrTime hi ha retard


DelayedFlights['Retard?'] = np.where(DelayedFlights['ArrTime'] > DelayedFlights['CRSArrTime'], 'Sí', 'No')

pd.options.display.max_columns = None

DelayedFlights

```

```python
# Salvem el nou Dataframe amb les noves columnes com a dataset "DelayedFlights_Reduit_Columnes.csv"

DelayedFlights.to_csv('DelayedFlights_Reduit_Columnes.csv')
```
